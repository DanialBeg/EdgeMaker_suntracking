```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
cap = cv2.VideoCapture(1)
   
while (1):
    _, img = cap.read()
    cv2.imshow('Original img',img)
    plt.show()
    (h, s, v) = cv2.split(img)
    s = s + 10;
    s = np.clip(s,0,255)
    img = cv2.merge([h,s,v])
    shadowc = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)
    edges = cv2.Canny(shadowc,100,200)
    cv2.imshow('Edges',edges)
    cv2.imshow('Converted Image',shadowc)
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([]), plt
 
    im = plt.imread('/Users/harris/Documents/FirstWebsite/Resources/sunset.jpg')
    plt.imshow(im,zorder=1, alpha=0.6)
    plt.show()
    plt.show()

    cv2.waitKey(0)
    cv2.destroyAllWindows()
 
```
